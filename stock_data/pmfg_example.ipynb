```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.read_gpickle('graph_info/2020-12.gpickle')
```

```python
plt.figure(figsize=(15,15))
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
nx.draw_networkx(G,with_labels=True,alpha = 0.8)
plt.title('PMFG in 202012')
plt.savefig('example: PMFG in 202012')
plt.show()
```

```python

degree_centrality=nx.degree_centrality(G)
degree_centrality_list= sorted(degree_centrality.items(),key=lambda x:x[1],reverse=True)
keys = [k[0] for k in degree_centrality_list]
values = [k[1] for k in degree_centrality_list] 
plt.figure(figsize=(25,8))
plt.title('Degree Centrality')

plt.xticks(rotation=90)
plt.plot(keys,values)
plt.savefig('Degree.png')
```

```python
betweenness_centrality=nx.betweenness_centrality(G)
betweenness_centrality_list= sorted(betweenness_centrality.items(),key=lambda x:x[1],reverse=True)
keys = [k[0] for k in betweenness_centrality_list]
values = [k[1] for k in betweenness_centrality_list] 
plt.figure(figsize=(25,8))
plt.title('Betweenness Centrality')
plt.xticks(rotation=90)
plt.plot(keys,values)
plt.savefig('betweenness.png')
```
